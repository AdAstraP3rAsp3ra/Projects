```python
# Import libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msno

# Read the data

df = pd.read_csv(r"C:\Users\user\OneDrive\Desktop\data\Movies Project\movies.csv")
```

```python
df.head()
```

```python
def summary(df):
    print(f"Dataset has {df.shape[1]} columns and {df.shape[0]} rows.")
    summary = pd.DataFrame(index=df.columns)
    summary["Unique"] = df.nunique().values
    summary["Missing"] = df.isnull().sum().values
    summary["Duplicated"] = df.duplicated().sum()
    summary["Types"] = df.dtypes
    return summary

summary(df)
```

```python
df = df.dropna()
```

```python
def summary(df):
    print(f"Dataset has {df.shape[1]} columns and {df.shape[0]} rows.")
    summary = pd.DataFrame(index=df.columns)
    summary["Unique"] = df.nunique().values
    summary["Missing"] = df.isnull().sum().values
    summary["Duplicated"] = df.duplicated().sum()
    summary["Types"] = df.dtypes
    return summary

summary(df)
```

```python
df["budget"] = df["budget"].astype("int64")
df["votes"] = df["votes"].astype("int64")
df["gross"] = df["gross"].astype("int64")
df["runtime"] = df["runtime"].astype("int64")
df.head()
```

```python
df.sort_values(by=["gross"],ascending = False)
```

```python
sns.regplot(data=df, x="budget", y="gross",line_kws= {"color":"red"})
plt.title("Budget vs Gross Earnings")
plt.xlabel("Film's Budget")
plt.show()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr(), annot = True)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Convert objects into numbers so i can see correlation
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if df_numerized[col_name].dtype == "object":
        df_numerized[col_name]= df_numerized[col_name].astype("category")
        df_numerized[col_name]= df_numerized[col_name].cat.codes

df_numerized
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
ax = sns.heatmap(df_numerized.corr(), annot = True)
plt.title("Correlation Matrix")
plt.show()
```
